```
import numpy as np
import pandas as pd
import keras

from keras.models import Sequential
from keras.layers import Dense,Flatten
from keras.layers import Dropout
from keras.layers.normalization import BatchNormalization

from keras.layers import Conv2D , MaxPooling2D
```

```
from google.colab import drive
drive.mount('/content/drive')
```

```
path = '/content/drive/My Drive/BTP/TrainImages.csv'
trainImages = pd.read_csv(path)
trainImages = trainImages.to_numpy()
```

```
path = '/content/drive/My Drive/BTP/TrainLabels.csv'
trainLabels = pd.read_csv(path)
trainLabels = trainLabels.to_numpy()

trainLabels = keras.utils.to_categorical(trainLabels,43)
```

```
cnn_trainImages = trainImages.reshape(len(trainImages), 30,30,1)

conv_Model = Sequential()

conv_Model.add(Conv2D(64, kernel_size=(3,3), activation='relu', input_shape=(30,30,1)))

conv_Model.add(Conv2D(128, kernel_size=(3,3), activation='relu'))
conv_Model.add(MaxPooling2D(pool_size=(2,2)))
conv_Model.add(BatchNormalization())

conv_Model.add(Conv2D(128, kernel_size=(3,3), activation='relu'))
conv_Model.add(MaxPooling2D(pool_size=(2,2)))
conv_Model.add(BatchNormalization())
conv_Model.add(Dropout(0.25))

conv_Model.add(Conv2D(128, kernel_size=(3,3), activation='relu'))
conv_Model.add(MaxPooling2D(pool_size=(2,2)))
conv_Model.add(BatchNormalization())
conv_Model.add(Dropout(0.5))

conv_Model.add(Flatten())

conv_Model.add(Dense(128,activation='relu'))

conv_Model.add(Dropout(0.5))

conv_Model.add(Dense(43, activation='softmax'))

conv_Model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
conv_Model.fit(cnn_trainImages, trainLabels, batch_size=128, epochs=100)

conv_Model.summary()
```

```
path = '/content/drive/My Drive/BTP/TestImages.csv'
testImages = pd.read_csv(path)
testImages = testImages.to_numpy()
```

```
path = '/content/drive/My Drive/BTP/TestLabels.csv'
testLabels = pd.read_csv(path)
testLabels = testLabels.to_numpy()

testLabels = keras.utils.to_categorical(testLabels,43)
```

```
cnn_testImages = testImages.reshape(len(testImages), 30,30,1)

_ , accuracy = conv_Model.evaluate(cnn_testImages , testLabels , verbose=0)
print('Accuracy: %.2f' % (accuracy*100))
```
